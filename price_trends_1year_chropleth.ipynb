```python
from urllib.request import urlopen
import plotly.express as px
import json
import numpy as np
import pandas as pd
import plotly.io as pio
pio.renderers.default = "browser"
```

```python
df = pd.read_csv((r'hpi_at_county.csv'),dtype={"FIPS code": str,"Annual Change (%)":float})
year = 2024
df = df[df['Year'] == year]

# log scale
df["Annual Change (signed log)"] = np.sign(df["Annual Change (%)"]) * \
                                   np.log10(1 + np.abs(df["Annual Change (%)"]))
maxi = df['Annual Change (signed log)'].max()
mini = df['Annual Change (signed log)'].min()
print(df.head(),'\n')

# where_is_Humboldt = df[df['FIPS code'] == '06023']
# print(where_is_Humboldt)
```

```python
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)

for i in range(len(counties["features"])):
    if counties["features"][i]['id'] == '06023':
        print (counties["features"][i])
```

```python
# fig = px.choropleth_map(df, geojson=counties, featureidkey='id', hover_name='County', locations='FIPS code', color='Annual Change (%)',
fig = px.choropleth_map(df, geojson=counties, featureidkey='id', hover_name='County', locations='FIPS code', color='Annual Change (signed log)',
                           # color_continuous_scale="Viridis",
                           color_continuous_scale="RdBu_r",
                        #    color_continuous_scale="Reds",
                           range_color=(mini, maxi),
                           map_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=0.5,
                           labels={'Increase':'Houce Price Increase %'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
