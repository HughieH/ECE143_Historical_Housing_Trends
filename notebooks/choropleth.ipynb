```python
from urllib.request import urlopen
import plotly.express as px
import json
import pandas as pd
```

```python
df = pd.read_excel("../data/raw/ccc.xlsx", dtype={"FIPS code": str})
print(df.head(),'\n')

print(df.tail(),'\n')
```

```python
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)

for i in range(len(counties["features"])):
    if counties["features"][i]['id'] == '06023':
        print (counties["features"][i])
```

```python
fig = px.choropleth_map(df, geojson=counties, featureidkey='id', hover_name='County', locations='FIPS code', color='HPI',
                           color_continuous_scale="Viridis",
                           range_color=(120, 220),
                           map_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=0.5,
                           labels={'HPI':'Houce Price Index'},
                           animation_frame = 'Year'
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
